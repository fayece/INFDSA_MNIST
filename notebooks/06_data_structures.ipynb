```python
from infdsa_mnist import dataset_loader, dast_calculations

(train_data, test_data) = dataset_loader.load_mnist()
x_train, y_train = train_data
x_test, y_test = test_data
```

In this notebook, we will explore how the dataset would act as different data structures and how this can affect the way we work with the dataset and train machine learning models. To that end, we will attempt to answer the following question:

To answer this, it is important to note that we need to take some creative liberties in translating the 2D arrays into different data structures, as the dataset is not structured to fit into all the data structures we explore. - `List`
- `Queue`
- `Dictionary`

## List
To explore how the dataset would act as a list, we can use the `to_list` function from the `dast_calculations` module. This function will take the training data as input and return a list. Each element in the list will be another list containing the pixel values of a single image and its corresponding label. This will allow us to see how the dataset can be interpreted as a list and how we could potentially work with it if we were to train a machine learning model using a list as the data structure.

```python
mnist_list = dast_calculations.to_list(x_train, y_train)

print(mnist_list[0])
```

> NOTE: Up to 4 [0]'s can be used to access the pixel values of the first image in the dataset.
> - The first [0] will give the first image in the list, including its label.
> - The second [0] will give the pixel values of the first image.
> - The third [0] will give the first row of the first image.
> - The fourth [0] will give the first pixel value of the first image. This illustrates how we can access different levels of information in the dataset using a list as the data structure.

The main advantages of using a list as the data structure for the dataset are:
- Simplicity: Lists are a simple data structure that many programmers are familiar with. This makes it easy to work with and understand, even for beginners.

Why lists are less than ideal:
- Flexibility: Lists are very flexible. However, this flexibility can lead to inconsistencies in data formatting and structure, which can quickly break the affected digits. It would also require the programmer to implement fail-safes to ensure that broken digits do not break the dataset. This will make the code more complex, harder to maintain, and more prone to bugs.
- Performance: Lists are not optimized for performance when it comes to large datasets. As the dataset grows in size, speed and memory usage can become significant bottlenecks, especially when performing operations that require complex data manipulation. This leads to slower training times and increased resource consumption.

## Queue
To explore how the dataset would act as a queue, we can use the `to_queue` function from the `dast_calculations` module. This function will take the training data as input and return a queue. Each element in the queue will be a tuple containing the pixel values of a single image and its corresponding label. This will allow us to see how the dataset can be interpreted as a queue and how we could potentially work with it, were we to train a machine learning model using a queue as the data structure.

```python
mnist_queue = dast_calculations.to_queue(x_train, y_train)

print(mnist_queue[0][0][0][0])
```

> NOTE: Up to 4 [0]'s can be used to access the pixel values of the first image in the dataset.
> - The first [0] will give the first image in the queue, including its label.
> - The second [0] will give the pixel values of the first image.
> - The third [0] will give the first row of the first image.
> - The fourth [0] will give the first pixel value of the first image.

The main advantages of using a queue as the data structure for the dataset are:
- Order: Queues maintain the order of the data, which can be beneficial when training machine learning models that require sequential data, such as recurrent neural networks (RNNs). This allows us to easily access the data in the order it was added, which can be important for certain types of models.
- Flexibility: Queues can be easily modified and extended, allowing for dynamic data processing and manipulation. This can be particularly useful when dealing with large datasets or when the data needs to be transformed in real-time during training.
- Efficiency: Queues can improve the efficiency of data processing by allowing for parallel processing and reducing memory usage. This can be especially important when working with large datasets or when training machine learning models on limited hardware resources.
- Scalability: Queues can handle large volumes of data and can be easily scaled to accommodate growing datasets. This makes them a good choice for training machine learning models on large datasets, as they can efficiently manage the data and ensure that it is processed in a timely manner.

The main disadvantages of using a queue as the data structure for the dataset are:
- Complexity: Queues can be more complex to implement and manage compared to other data structures, especially for those who are not familiar with the concept of queues. This can lead to increased development time and potential for bugs if not implemented correctly.
- Performance: While queues can improve efficiency in certain scenarios, they may not always be the most efficient data structure for all types of operations. For example, if the dataset requires access to specific elements that aren't in order, it may lead to increased time complexity and slower performance compared to lists or dictionaries.

## Dictionaries
To explore how the dataset would act as a dictionary, we can use the `to_dict` function from the `dast_calculations` module. This function will take the training data as input and return a dictionary. Each key in the dictionary will be the label of a single image, and the corresponding value will be a list containing the pixel values of that image. This will allow us to see how the dataset can be interpreted as a dictionary and how we could potentially work with it, were we to train a machine learning model using a dictionary as the data structure.

```python
mnist_dict = dast_calculations.to_dict(x_train, y_train)

print(mnist_dict[0][0][0][0])
```

> Note: There are 4 [0]'s that can be used to access the pixel values of the first image in the dataset.
> - The first [0] will select the label of the image that we want to access. In the initial example, we are accessing the label '0.'
> - The second [0] will select the first image in the selected label.
> - The third [0] will select the first row of the selected image.
> - The fourth [0] will select the first pixel value of the selected image.

The main advantages of using a dictionary as the data structure for the dataset are:
- Fast lookups: Dictionaries provide constant time complexity for lookups, which can be beneficial when training machine learning models that require frequent access to specific data points. This allows us to quickly retrieve the pixel values of a specific image based on its label, which can be useful for certain types of models.
- Flexibility: Dictionaries can be easily modified and extended, allowing for dynamic data processing and manipulation. This can be particularly useful when dealing with large datasets or when the data needs to be transformed in real-time during training.
- Organization: Dictionaries can help organize the data in a way that is more intuitive and easier to understand. By using labels as keys, we can easily group images by their corresponding labels, which can be beneficial for certain types of models that require class-based data organization.

The main disadvantages of using a dictionary as the data structure for the dataset are:
- Memory usage: Dictionaries can consume more memory than other data structures, especially when dealing with large datasets. This can be a concern when training machine learning models on limited hardware resources, as it may lead to increased memory usage and slower performance.
- Complexity: Dictionaries can be more complex to implement and manage compared to other data structures, especially for those who are not familiar with the concept of dictionaries. This can lead to increased development time and potential for bugs if not implemented correctly. Additionally, if the dataset contains a large number of unique labels, the dictionary may become more difficult to manage and may require additional handling to ensure that it remains efficient and organized.
- Performance: Dictionaries can have slower performance compared to other data structures, such as lists or arrays, when it comes to accessing elements by index. This can be a concern when dealing with large datasets or when the data needs to be accessed frequently during training. Additionally, dictionaries may have slower performance when it comes to iterating over the dataset, as they require additional overhead to access the values associated with each key.

## Linear vs Non-linear data structures





The data structures we have explored so far (lists, queues, and dictionaries) are either linear or non-linear. Linear data structures store data sequentially, while non-linear data structures store data in a more complex structure. Following are some examples of both:
#### Linear data structures
- Array
- List
- Vector
- Matrix
- Linked list
- Queue

#### Non-linear data structures
- Dictionary
- Tree
- Graph

### Testing linear vs non-linear data structures
Instead of testing the performance of linear vs non-linear data structures, we will instead explore which data structure types are more suitable for different tasks.

#### Saving a single MNIST image as an array of pixel values
The most suitable data structure type for saving a single MNIST image as an array of pixel values would be a linear data structure. This is because saving a single image is a simple task that does not require the complexity of a non-linear data structure. An array specifically would be a great choice, as it allows for efficient storage, low memory overhead, and fast access to pixel values.

#### Saving all MNIST images in a dataset
The most suitable data structure type for saving all MNIST images in a dataset would be a non-linear data structure. This is because saving all images in a dataset requires a more complex structure if we want the data to be organized in an easy-to-manage manner. A tree would be a great choice, as it allows for efficient storage and organization of the images based on their labels and other attributes. This would allow for easy access and reduced memory usage.

#### Counting the number of each digit (0–9) in the dataset
The most suitable data structure type for saving all MNIST images in a dataset would be a non-linear data structure. As explained in the previous tasks, this is due to the ability to organize the data based on the labels, which allows for O(1) time complexity when counting the number of each digit in the dataset. A dictionary would be a great choice for this task, as its only requirement is to have the labels as keys. This makes the task trivially easy.

#### Grouping all images of the same digit together
The most suitable data structure type for grouping all images of the same digit together would be a non-linear data structure, for all the same reasons as counting the number of each digit. A dictionary in particular would be a great choice, since it stores all images of the same digit together and could easily count them as well.

#### Using a decision tree classificator to classify all digits in the dataset
The most suitable data structure type for using a decision tree classifier to classify all digits in the dataset would be a non-linear data structure. This is because decision trees are a type of non-linear model that can capture complex relationships between features and labels. A tree data structure would be a great choice for this task, as it allows for efficient storage and organization of the data based on the features and labels. This would allow for easy access and reduced memory usage when training the decision tree classifier on the dataset. Additionally, the tree structure would allow for efficient traversal and decision-making during the classification process, which is essential for the performance of the decision tree classifier.
