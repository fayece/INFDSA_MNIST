```python
from infdsa_mnist import vian_display, dataset_loader

(train_data, test_data) = dataset_loader.load_mnist()
x_train, y_train = train_data
x_test, y_test = test_data
```

In this notebook, we will perform a visual analysis of the MNIST dataset. To that end, we will attempt to answer the following question:
**Why is classification of handwritten digits a difficult task?**

To answer this question, we will explore the dataset visually and identify potential challenges that may arise when trying to classify handwritten digits.

We must understand how the digits are visually represented in the dataset to identify potential challenges in classification. To explore the data visually in a structured manner, we will subdivide the main question into several sub-questions that will guide our visual analysis of the dataset. These sub-questions will aide us in identifying specific visual characteristics of the digits that may pose challenges for classification. The sub-questions we will address in this notebook are as follows:

1. How can multiple handwritten digits of the same digit class differ visually whilst still being recognizable as that digit?
2. Why are certain digit classes more challenging to distinguish from one another based on visual features alone?

## How can multiple handwritten digits of the same digit class differ visually whilst still being recognizable as that digit?
To explore the visual differences between multiple handwritten digits of the same class, we can use the `display_single_digit` function from the `vian_display` module. This function allows us to display multiple examples of a specific digit class, enabling us to visually analyze the variations in handwriting styles while still recognizing them as the same digit.

```python
digit = 7
amount = 10
seed = 42
vian_display.display_single_digit(x_train, y_train, digit, amount, seed)
```

The displayed images of the digit '7' show significant visual differences in handwriting styles, yet are still recognizable as the same digit. This illustrates the inherent variability in how individuals write the same digit, which can pose challenges for classification.
To break this down, the digit '7' can largely be broken down into the following elements:
- At the top: A line. Shape options: straight horizontal line, tilde-shaped (~), or slightly diagonal.
- At the top-left: An optional short "hook" vertical line.
- Downwards from the top-right: A Line. Shape options: straight vertical, slightly diagonal, curved.
- In the middle of the downwards line: An optional line. Shape options: straight horizonal, tilde-shaped (~), or slightly diagonal.

This breakdown of the digit '7' into its elements and variations in their shapes illustrates how cultural and individual handwriting styles can lead to significant visual differences within the same digit class. This makes it challenging for classification algorithms to accurately identify the digit '7' across different handwriting styles, and must be taken into account when designing a model that can classify handwritten digits effectively.
Although the digit '7' was used as an example here, every digit class (0–9) can exhibit similar variability in handwriting styles, which contributes to the overall challenge of classifying handwritten digits accurately.

## Why are certain digit classes more challenging to distinguish from one another based on visual features alone?
To explore why certain digit classes may be more challenging to distinguish from one another based on visual features alone, we can use the `display_multiple_digits` function from the `vian_display` module. This function allows us to display examples of multiple digit classes side by side, enabling us to visually analyze the similarities and differences between them and identify potential challenges in distinguishing them based on visual features alone.

```python
samples_per_digit = 5
similar_digits = [
    (1, 7),
    (5, 6),
    (2, 7),
    (9, 4),
    (0, 6)
]

subset_size = 1000

vian_display.display_similar_digits(x_train, y_train, samples_per_digit=samples_per_digit, similar_digits=similar_digits)
```

> NOTE: The displayed images are compared using a Euclidean similarity measure. This means that the images are compared based on their pixel values, and the similarity is calculated based on the distance between those pixel values. It approximates visual similarity and illustrates a potential challenge in distinguishing between digit classes using a computed process rather than human perception.

In the displayed images, we can see that certain digit classes may have similar visual features that can make them challenging to distinguish from one another. Aside from this, as there is a multitude of handwriting styles and thicknesses that can be applied to the same digit, it can lead to significant visual overlap between different digit classes.

Some observations relating to visual overlap we can make based on the displayed images are:
- Stroke thickness (thick or thin lines)
- Handwriting style (cursive, mimicking printed digits, etc.)
- Start and end points of strokes (short or long lines)
- Flow of strokes (sharp or curved lines)
- Presence/absence of certain elements (extra lines that may be present in some versions of the digit)
- Meticulousness of handwriting (whether a digit is written sloppily or accurately)

These observations illustrate how certain digit classes can have similar visual features that can make them challenging to distinguish from one another based on visual features alone. This is an important consideration when designing a model for classifying handwritten digits. The model must be able to distinguish between digit classes with similar visual features based on other characteristics that can't be easily parsed. It must be robust enough to see past the visual similarities and identify the underlying digit class accurately without relying solely on brittle connections between visual features and digit classes.

## Conclusion
By performing a visual analysis of the MNIST dataset, we have identified several hurdles that must be overcome to classify handwritten digits accurately. These hurdles are as follows:
- Variability in handwriting styles within the same digit class. As seen with the digit '7,' there can be significant visual differences in how individuals write the same digit. Stroke length, angles, additions, and general handwriting styles can change the makeup of the digit significantly. This means a model cannot rely on a single template to recognize and categorize a digit.
- Similarities between different digit classes. When comparing digit classes side by side, such as '1' and '7,' or '4' and '9,' we can see that certain digit classes can have a very similar visual appearance, especially when considering the variability in handwriting styles.

These two hurdles combined create ambiguity in the visual features of the digits, making it difficult for a model to accurately classify them based on visual features alone. An unclosed top in a '9' due to sloppy writing can make it look like a '4.' This highlights the importance of designing a model that can "see" beyond the visual features. The model should recognize the structure and underlying characteristics of the digits, rather than relying solely on visual features that can be easily confused due to handwriting variability.
