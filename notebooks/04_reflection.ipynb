## What information does a human use to recognize digits?
Unfortunately, finding a definitive answer to this question is not possible, as the way humans recognize digits is very complex and not fully understood, even by neuroscientists and other specialists. If we were to try to answer this question, we would need to conduct a thorough study of the human brain and its processes. This would be a very time-consuming and expensive task, especially when this sub-question is not the main focus of the notebook. Therefore, we will focus on the broader question of how humans recognize objects in general and make assumptions about how humans recognize digits, based on the information we have gathered.

Humans appear to use a mathematical system to organize visual objects based on their principal components, that is to say, the most important features that define the object. The brain maintains something akin to a 2D map where objects with similar features are represented near one another. Take, for example, faces and apples. Since they are both curvy and round, they are grouped together. On the other hand, a face is not at all similar to a spiky object like a helicopter or chair, so these are grouped far apart on the map. [1]

Although digits weren't specifically studied in the research mentioned in the article, we can assume that the brain uses the same principles to organize digits. If we were to draw a potential map for digits, we would see that these are the four main groups, along with their corresponding numbers:
- Symmetrical and curvy: 0, 8
- Symmetrical and straight: 1
- Asymmetrical and curvy: 2, 3, 5, 6, 9
- Asymmetrical and straight: 4, 7

This map suggests the brain could recognize digits almost immediately. However, the largest group, "Asymmetrical and curvy," contains half of the digits, implying that a simple 2D map would not be enough to distinguish them. Therefore, the brain likely relies on a more complex system, taking into account possible variations, stroke patterns, prior encounters with different versions of the same digits, and other factors. These groupings are merely simplified assumptions to illustrate how the brain might begin the recognition process.

## What information can a computer receive for the same task?
In the previous notebooks we have explored the MNIST dataset and have seen that a computer receives images in the form of 28x28 grayscale images, where each pixel value is an integer between 0 and 255. This means that a computer, per image, receives a 2D array of values between 0 and 255. The values represent the intensity of the pixel, with 0 being black and 255 being white. A computer can use this information to try to recognize the digits in the images.

A computer, unlike a human, does not inherently know which patterns or features are important for recognizing digits. It must rely on the information it receives from the dataset to try to recognize the digits. This means a computer must learn to recognize the digits without prior knowledge of what the digits look like.

To use an analogy, this is similar to trying to teach a child what the color red is. A child will not know the color red beforehand, but they will learn to recognize the color red based on the examples they see. However, this does not take into consideration that there are many different colors that are red, and that a wine red shirt and a red apple are both red, despite looking different.
Similarly, a computer must learn to recognize that a '7' can look different based on handwriting styles, stroke thickness, cultural differences, and other factors. It must learn the concept of '7' rather than relying on the first version it sees in the dataset.

To further expand on this, if a '7' were to be shifted a single pixel to the left, it would look very different to a computer as the array of pixel values would change significantly. However, a human would still have no trouble recognizing it as a '7' since humans are not reliant on contextual pixel values to recognize digits.

## Are pixels enough for a computer to recognize digits?
As we have seen in `02_visual_analysis`, there are many different ways that a digit can be written, and handwriting styles can vary greatly. The examples shown in the notebook show how visually similar different digits can appear. If a computer relied solely on pixel values, recognizing digits accurately would be challenging due to such variability.


This is why the MNIST dataset has labeled each digit. A machine learning model can learn to associate certain pixel patterns with the correct digit classes, allowing it to recognize digits even when their appearance does not match the examples it sees in the training data. However, the model would still need to generalize from the training examples to correctly recognize unseen variations of the digits. Therefore, effective generalization is necessary for accurate digit recognition.

## Why is MNIST a popular dataset as a practice dataset for machine learning?
The MNIST dataset is a popular dataset for practicing machine learning for several reasons:
1. The dataset is relatively small, which makes it computationally efficient to work with. This allows for quick experimentation and iteration for beginners [2].
2. The dataset is standardized, with a fixed size of grayscale 28x28 images, all centered and normalized. This makes it easier for beginners to focus on learning machine learning concepts rather than dealing with the complexities of data preprocessing and cleaning [3].
3. The dataset is publicly available, making it accessible for anyone to use and experiment with machine learning models.

## Conclusion
By reflecting on the differences between how humans and computers recognize digits, we can now answer the main question posed in the introduction: **How can we use machine learning to solve this human-level task?**

To use machine learning to solve the human-level task of digit recognition, we need to bridge the gap between how a human brain processes visual information and how a computer processes raw numerical arrays.

From our reflection, we can summarize the approach as follows:
- A human brain naturally extracts features (curves, symmetry, etc.) and maps them to concepts based on prior experience. It can ignore shifts in position, scaling, messy handwriting, and other variations that do not change the underlying concept of the digit.
- A computer does not receive any conceptual information. It only sees a grid of pixel values. A single pixel shift can fundamentally change the input data, making it look completely different to the computer, even though it may look the same to a human.

Therefore, raw pixels are not enough for a computer to recognize digits. We must use a supervised learning approach, where we provide the computer with labels for each image in the dataset. This allows the computer to learn to associate certain pixel patterns with the correct digit classes.
Because the MNIST dataset already comes pre-configured with labeled data, it is already a great starting point for practicing machine learning. Add to that the fact the dataset is relatively small and the images standardized, and it becomes clear why MNIST has become as popular as it is for its intended purpose.

Though we understand the challenges of the task and how we can use machine learning to solve it, we must acknowledge that there may be challenges in the data quality that can affect the performance of machine learning models.
